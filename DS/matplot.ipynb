```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("x vs x^2 graph")
# plt.show()

#can play around with the line
plt.plot(x,y,color="red",linestyle="-.",marker="*",linewidth = 3, markersize = 9)
plt.grid()
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
z = np.array([1,2,3,4,5])

#size of plot default.
plt.figure(figsize=(9,5))#The figsize argument sets the width and height of the figure in inches. In this case, the figure will be 9 inches wide and 5 inches tall.

plt.subplot(2,2,1)
plt.plot(x*2,y,color = "red")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x**2,y,color = "blue")

plt.subplot(2,2,3)
plt.plot(y,z,color = "green")

plt.subplot(2,2,4)
plt.bar(y,y,color = "black")

```

```python
#histogram ; create a distibution of number of data points

data = [1,2,2,2,2,3,3,4,4,4,5]
plt.hist(data,bins=5,color='orange',edgecolor='black')

```

```python
#scatter plot
plt.scatter(x,y,color="blue",marker="x")
```

```python
# Pie Chart

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
explode = (0.1, 0, 0, 0)  # explode first slice

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Pie Chart Example")
```

```python
#Sales data visualisation

import pandas as pd

sales_data_df = pd.read_csv(r"C:\Users\abhin\OneDrive\Documents\code\ML\Udemy\data\sales_data.csv")
sales_data_df.head(5)
```

```python
groupby_sales = sales_data_df.groupby(["Product Name"])["Units Sold"].sum()
groupby_sales
```

```python
groupby_sales = sales_data_df.groupby(["Product Category"])["Total Revenue"].sum()
groupby_sales
```

```python
#plot bar on these sales
groupby_sales.plot(kind="bar",color="teal")
```

```python
#sales trend over time and reset index to get the format right
sales_trend = sales_data_df.groupby(["Date"])["Total Revenue"].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend["Date"][:15], sales_trend["Total Revenue"][:15])
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.title("Sales Trend - First 7 Days")
plt.xticks(rotation=45)
plt.tight_layout()
```
