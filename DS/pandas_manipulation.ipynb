```python
BASE_DIR = os.getcwd()
print(BASE_DIR)
csv_path = "./data/data.csv"
```

```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv(csv_path)
```

```python
df.head(5)
```

```python
#Handling missing values 
df.isnull() #shows all the missing values with True 
df.isnull().any(axis=1)  # This returns a boolean Series indicating whether each row contains at least one missing value (True if any column in the row is NaN, otherwise False).
df.isnull().sum()#show the number of missing
```

```python
df_filled = df.fillna(0)
#fill missing values with the mean of the column
df["Sales_fillNA"] = df['Sales'].fillna(df["Sales"].mean())
df.head(5)
```

```python
#change datatype of a specific col and rename col
df = df.rename(columns = {"Date":"Sales Date"})
df.head(5)
#create new column
df['Value_New'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head(5)
```

```python
df['New Value'] = df['Value'].apply(lambda x:x*2)
```

```python
#data aggragating and grouping? ->basically means to group a category and do some aggragation like sum()mean()etc 
grouped_mean = df.groupby("Product")["Value"].mean()
print(grouped_mean)
```

```python
#group both the categories
group_by_sum = df.groupby(["Product","Region"])["Value"].sum()
print(group_by_sum)
```

```python
group_by_agg = df.groupby(['Product',"Region"])["Value"].agg(["mean","sum","count"])
group_by_agg
```

```python
df1 = pd.DataFrame([{"Key":"A","Value":"1"},{"Key":"B","Value":"3"},{"Key":"C","Value":"9"}])
df2 = pd.DataFrame([{"Key":"A","Value":"2"},{"Key":"B","Value":"6"},{"Key":"D","Value":"8"}])
```

```python
#Merging dfs #ven diagram
pd.merge(df1,df2,on="Key",how="inner")#intersection
pd.merge(df1,df2,on="Key",how="outer")#union
pd.merge(df1,df2,on="Key",how="left")#A - B
pd.merge(df1,df2,on="Key",how="right")#B - A
```
