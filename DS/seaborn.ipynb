```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
import matplotlib.pyplot as plt
#no x label and y label
sns.scatterplot(x = "total_bill",y = "tip", data = tips)
plt.title("Scatter Plot of total bill vs tip")
```

```python
##line plot
#See how it is visually better than matplotlib 
sns.lineplot(x = 'size', y = 'total_bill', data=tips)
plt.title("Total bill by size")
```

```python
#Bar plot
sns.barplot(x = 'day', y = 'total_bill', data=tips)
plt.title("Total bill by day bar")
```

```python
#box plot
sns.boxplot(x = "day", y ="total_bill", data = tips )
#box plot has outliers like min and max, sat has most min bill and sun has max bill
```

```python
#violin plot
#what is this lol 
sns.violinplot(x = "day", y ="total_bill", data = tips )
```

```python
#histograph
sns.violinplot(x = "day", y ="total_bill", data = tips )
```

```python
sns.histplot(tips["total_bill"],bins = 10, kde=True)#kernel density estimator.
```

```python
sns.kdeplot(tips["total_bill"], fill = True)
```

```python
#pairplot - relation bw all the dependent variables
sns.pairplot(tips)#works only on numerical data
```

```python
#CORRELATION BETWEEN NUMERICAL VALUES
corr = tips[["total_bill","tip","size"]].corr()
corr#shows positive correlation and we can plot with heat maps
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")#annot shows number, and the other one is coloe
```
