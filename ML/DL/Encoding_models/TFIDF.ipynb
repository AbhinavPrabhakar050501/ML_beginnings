```python
import pandas as pd
messages = pd.read_csv("SMSSpamCollection.txt", sep="\t", names=["label","message"])
```

```python
messages
```

```python
#preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer() 

```

```python
corpus = []
for i in range(0,len(messages)):
    sentence = re.sub('[^a-zA-z]',' ',messages['message'][i])#remove special characters
    sentence = sentence.lower()#lower case
    sentence = sentence.split()#convert to list
    #stemmization for root word
    sentence = [lemmatizer.lemmatize(word) for word in sentence if word not in stopwords.words('english')]
    sentence = ' '.join(sentence)
    corpus.append(sentence)

    






```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=100)#choose the top 100 features
```

```python
X = cv.fit_transform(corpus).toarray()
```
