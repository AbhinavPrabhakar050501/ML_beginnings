```python
from nltk.stem import PorterStemmer as p_stem
from nltk.stem import RegexpStemmer as r_stem
from nltk.stem import SnowballStemmer as s_stem

porter = p_stem()
porter.stem("fairly")
```

```python
porter = r_stem('ing$|s$|e$|able$|ly$')
porter.stem("fairly")
```

```python
porter = s_stem("english")
porter.stem("fairly")
```
