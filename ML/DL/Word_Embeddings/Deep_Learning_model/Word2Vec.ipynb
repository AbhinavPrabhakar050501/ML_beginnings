```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
import gensim.downloader as api
w2v_model = api.load('word2vec-google-news-300')
```

```python
king_vector = w2v_model['king']
print(king_vector.shape)#300 dim
```

```python
similar_to_cricket = w2v_model.most_similar('cricket')
print(similar_to_cricket)
```

```python
similarity_score = w2v_model.similarity('hockey', 'sports')
print(similarity_score)
```

```python
result_vector = w2v_model['king'] - w2v_model['man'] + w2v_model['woman']
```

```python
similar_words = w2v_model.similar_by_vector(result_vector)
print(similar_words)
```
