```python
import pandas as pd
messages = pd.read_csv("SMSSpamCollection.txt", sep="\t", names=["label","message"])
```

```python
messages
```

```python
#preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
ss = SnowballStemmer("english"
)
```

```python
corpus = []
for i in range(0,len(messages)):
    sentence = re.sub('[^a-zA-z]',' ',messages['message'][i])#remove special characters
    sentence = sentence.lower()#lower case
    sentence = sentence.split()#convert to list
    #stemmization for root word
    sentence = [ss.stem(word) for word in sentence if word not in stopwords.words('english')]
    sentence = ' '.join(sentence)
    corpus.append(sentence)

    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True, max_features=100)#choose the top 100 features
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
cv.vocabulary_
```

N_GRAMS

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))#adding N_grams for better differentiation \
X = cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_ #shows the N_gram words here 2 words so bigram only.  
# also to note that numbers are the index in the column, lower the number
# higher the frequency as it frequency of occurences descending
```

```python
y = pd.get_dummies(messages['label'])
```

```python
y 
```

```python
y = y.iloc[:,0].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred)
```

```python
print(classification_report(y_true=y_test,y_pred=y_pred))
```
