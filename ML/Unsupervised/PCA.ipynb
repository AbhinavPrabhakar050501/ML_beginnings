```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df = pd.DataFrame(cancer_dataset['data'], columns= cancer_dataset['feature_names'])
```

```python
df
```

```python
#standard scaling for feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
# APPLYIYING PCA ALGORITHM
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
dim_red_data = pca.fit_transform(scaled_data)
#Dimensionality reduction
```

```python
pca.explained_variance_ #based on the n number of features
```

```python
# Scatter plot of the two principal components colored by class
pc1 = dim_red_data[:, 0]
pc2 = dim_red_data[:, 1]
targets = cancer_dataset['target']
target_names = cancer_dataset['target_names']

plt.figure(figsize=(8, 6))
plt.scatter(pc1,pc2, c = cancer_dataset['target'], cmap = 'viridis', edgecolor='k', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
```
